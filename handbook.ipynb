1. 查询某国/地区的指标数据（所有可查年份）
```python
data = Country('国名/地区名').fetch_data('指标名', '年份')
```

### 2.1 查询某一年的指标
比如，中国2018年的GDP数据为：

````python
data = Country('China').query('GDP', '2018')
data
````

转换成`DataFrame`格式为：

````python
pd.DataFrame(data)
````

### 2.2 查询所有可查年份的指标

比如，中国可查的GDP数据：

````python
data = Country('China').fetch_data('GDP')
pd.DataFrame(data)
````

## 4. 多线程获取所有国家/地区的某指标数据

````python
from knoema.knoema import fetch_indicator_of_all_countries
````

````python
fetch_indicator_of_all_countries('GDP')
````

此时，会生成数据文件`./GDP_of_all_countries.json`，可以通过`load_data_file_to_dataframe`方法来载入数据。

````python
df = load_data_file_to_dataframe('./GDP_of_all_countries.json')
df.head(10)
````

## 5. 多线程获取某国家/地区的所有指标数据

````python
from knoema.knoema import fetch_all_indicators_of_country
````

````python
fetch_all_indicators_of_country('China')
````
